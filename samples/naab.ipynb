```python
from datasets import load_dataset, Dataset, DatasetDict
from datasets import load_dataset, Dataset, DatasetDict, load_from_disk


dataset_stream = load_dataset(
    "text",
    data_files={
        "train": [
            "https://huggingface.co/datasets/SLPL/naab/resolve/main/data/train-00005-of-00126.txt",
            "https://huggingface.co/datasets/SLPL/naab/resolve/main/data/train-00001-of-00126.txt",
        ],
        "test": [
            "https://huggingface.co/datasets/SLPL/naab/resolve/main/data/test-00000-of-00003.txt",
        ],
    },
    streaming=True,
)

# take only small subsets
train_small = list(dataset_stream["train"].take(2000))  # adjust size
test_small = list(dataset_stream["test"].take(500))

# convert to regular Dataset objects
train_ds = Dataset.from_list(train_small)
test_ds = Dataset.from_list(test_small)



# combine train and test
dataset_small = DatasetDict({"train": train_ds, "test": test_ds})

# save to disk
dataset_small.save_to_disk(
    "/mount/studenten-temp1/users/setare/semantic_change_database/naab/naab_subset_small"
)


dataset_small = load_from_disk(
    "/mount/studenten-temp1/users/setare/semantic_change_database/naab/naab_subset_small"
)

print(dataset_small)
print(dataset_small["train"][0])
```
